### Data Loading

```python
def load_pair(pair):
    df = pd.read_csv(f"../raw/{pair}_USDT_{origin[pair]}_1h.csv",index_col=0,parse_dates=True)    
    df.rename(columns = {'volumefrom':'volume', 'volumeto':'USDT_volume'}, inplace = True)

    return df

def plot_ohcl_data(pair):
    ohcl_data = load_pair(pair)
    ohcl_data.index.name = 'Date'
    
    # For candle plots
    #mc = mpf.make_marketcolors(up='g',down='r')
    #s  = mpf.make_mpf_style(marketcolors=mc)
    
    #mpf.plot(ohcl_data, type='line', style='charles', datetime_format='%d-%m-%Y', title=f"{pair}", ylabel='Price (USDT)',
    #         tight_layout=False, figratio=(10,5), volume=False)
    plt.plot(ohcl_data["close"], lw=1.5, label="Closing Price")
    plt.xlabel("Year", fontweight="bold", fontsize=18)
    plt.ylabel("USDT", fontweight="bold", fontsize=18)
    plt.title(pair, fontweight="bold")
    plt.tight_layout()
    plt.legend()
    #if pair == "BTC":
    #plt.savefig(f"{pair}.png", dpi=600)
    plt.show()
```

```python
for pair in origin.keys():
    if pair == "BTC":
        #plot_ohcl_data(pair)
        ohcl_data = load_pair(pair)
        ohcl_data.index.name = 'Date'

        plt.figure(figsize=(12,7))
        plt.grid(True, linestyle="dotted", linewidth=2.5)
        plt.plot(ohcl_data["close"], lw=1.5, alpha=1, label="Closing Price")
        plt.xlabel("Year", fontweight="bold", fontsize=16)
        plt.ylabel("USDT", fontweight="bold", fontsize=16)
        plt.title(pair, fontweight="bold", fontsize=18)
        plt.tight_layout()
        plt.legend(loc="upper left")
        #if pair == "BTC":
        plt.savefig(f"{pair}_1.png", dpi=600)
        plt.show()
```
