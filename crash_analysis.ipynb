```python
from cryptoportfolio.data.portfolio import Portfolio

year = "2022"
pf = Portfolio(year, synthetic=False, split="whole")#, recycle=False, overwrite=True)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from sstudentt import SST

import seaborn as sns
sns.set_style('darkgrid')
sns.set(color_codes=True)
sns.set_context('talk', font_scale=1)
sns.set_palette("rocket", len(pf._portfolio))

plt.figure(figsize=(14,8))   
plt.grid(True, linestyle="dotted", linewidth=2.5)
plt.title(f"{year} Crash - Portfolio Values", fontweight="bold", fontsize=20)

lw = 3.5; alpha = 0.6
for key, value in pf._portfolio.items():  

    #plt.figure(figsize=(14,8))   
    #plt.grid(True, linestyle="dotted", linewidth=2.5)

    if key == "BTC":
        plt.plot(value["close"]/value["close"][0], label=key, lw=5, alpha=1)
    else:
        plt.plot(value["close"]/value["close"][0], label=key, lw=lw, alpha=alpha)
    plt.ylabel("Relative USDT Price", fontweight="bold", fontsize=16)
    plt.xlabel("Date", fontweight="bold", fontsize=16)
    #plt.xlabel("Trading Period")
    plt.legend(prop={'size': 13})
    plt.tight_layout()
    plt.yscale('log')
    if year == "2018":
        plt.yticks([5, 1, 0.5, 0.2, 0.1, 0.05], ['$5 x 10^0$', '$10^0$', '$5x10^{-1}$', '$2x10^{-1}$', '$10^{-1}$', '$5x10^{-2}$'])
    elif year == "2021":
        plt.yticks([5, 1, 0.5, 0.4, 0.3, 0.2], ['$5 x 10^0$', '$10^0$', '$5x10^{-1}$', '$4x10^{-1}$', '$3x10^{-1}$', '$2x10^{-1}$'])
    elif year == "2022":
        plt.yticks([2, 1, 0.5, 0.2, 0.1], ['$2 x 10^0$', '$10^0$', '$5x10^{-1}$', '$2x10^{-1}$', '$10^{-1}$'])
    #plt.ylim([0.1,1.1])
#plt.savefig(f"crash_{year}.png", dpi=600)
plt.show()
```
